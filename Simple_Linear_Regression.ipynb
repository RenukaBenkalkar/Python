```python
#Simple Linear Regression using Advertising SlR Dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#load dataset
from google.colab import files
d=files.upload()
```

```python
#Read Dataset
df =pd.read_csv('advertising SLR.csv')
df.head()
```

```python
#Regression Algorithms
#x is TV and Y is sales
x=df.iloc[:,0]
y=df.iloc[:,-1]

x.head()
y.head()
```

```python
#Train-Test split

from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,train_size=0.7,random_state=0)

```

```python
xtrain.shape   #70% of data out of 200 is taken for training purpose
```

```python
xtest.shape     #30% of data is talen for testing
```

```python
xtrain
```

```python
#Build Simple Linear Algorithms
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()

#lr.fit(xtrain,ytrain)-- error convert single dimension to 2 dimension
```

```python
#convert data from single dimension to two dimension in SLR

xtrain=xtrain.values.reshape(-1,1)
xtest=xtest.values.reshape(-1,1)
ytrain=ytrain.values.reshape(-1,1)
ytest=ytest.values.reshape(-1,1)
```

```python
xtrain.shape   #converted to 2 dimension
```

```python
ytest
```

```python
lr.fit(xtrain,ytrain) # model Building
```

```python
lr.fit(xtest,ytest)
```

```python
#Model prediction
ypred = lr.predict(xtest)
ypred
```

```python
# Residual errors
err =ytest -ypred
err
```

```python
# Visualization
plt.scatter(xtest,ytest,color='blue') #actual data
plt.plot(xtest,ypred,color='red')     #predicted data
plt.title('Linear Regression')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```

```python
# Evaluation matrix - Checking the performance of the model

from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score,root_mean_squared_error
```

```python
mean_absolute_error(ytest,ypred)
```

```python
mean_squared_error(ytest,ypred)
```

```python
root_mean_squared_error(ytest,ypred)
```

```python
r2_score(ytest,ypred) # SLR predicts result 81%
```
